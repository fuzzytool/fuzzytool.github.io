# Fuzzy clustering

Cluster synthetic blobs with FCM, inspect memberships, and pick the number of clusters with a validity index.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import fuzzytool as fz
from fuzzytool import cluster, viz
from fuzzytool.datasets import make_blobs
```

```python
X = make_blobs(centers=((0, 0), (6, 6), (0, 6)), seed=0)
res = fz.fuzzy_cmeans(X, c=3, seed=0)
viz.plot_clusters(X, res); plt.show()
```

## How many clusters?

Sweep `c` and pick the partition coefficient's maximum.

```python
for c in range(2, 7):
    r = fz.fuzzy_cmeans(X, c=c, seed=0)
    pc = cluster.partition_coefficient(r.u)
    xb = cluster.xie_beni(X, r.centers, r.u)
    print(f'c={c}: PC={pc:.3f}  XB={xb:.3f}')
```
